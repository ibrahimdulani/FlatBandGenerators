```python
import os.path as osp

import numpy as np
import networkx as nx
import scipy.linalg as la
import scipy.sparse as ssp
import matplotlib.pyplot as plt
import seaborn as sb

import lsm
```

```python
n = 6
s3 = np.sqrt(3)
uc = np.array([[1.5,0],[0.5,0],[0,s3/2],[0.5,s3],[1.5,s3]])
lpt= np.zeros((5*n,2))
for i in range(n):
    for j in range(5):
        lpt[5*i+j,0] = uc[j,0]+2*i
        lpt[5*i+j,1] = uc[j,1]
 
```

```python
x = lpt[:,0]
y = lpt[:,1]
xx = np.append(x,x[len(x)-5]+0.5)
yy = np.append(y,s3/2)
plt.axes().set_aspect( 'equal')
plt.axes().set_ylim(-0.3,2)
plt.scatter(xx,yy)

for i in range(len(x)):
    if i!=0 and i!=len(x)-1 and i%5!=0 and i%5!=4:
        plt.plot(x[[i-1,i,i+1]],y[[i-1,i,i+1]])
    if i%5==0:
        if i==0:
            plt.plot(x[[i,i+7]],y[[i,i+7]])
        elif i!=len(x)-5:
            plt.plot(x[[i,i+7]],y[[i,i+7]])
    elif i%5==4:
        if i!=len(x)-1:
            plt.plot(x[[i,i+3]],y[[i,i+3]])
plt.plot([x[len(x)-5],x[len(x)-5]+0.5,x[len(x)-1]],[y[len(x)-5],s3/2,y[len(x)-1]])
     

plt.show()
```
